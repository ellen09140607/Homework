#### Read the dataset in the csv file from the URL


```python
from js import fetch
import io

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ST0151EN-SkillsNetwork/labs/boston_housing.csv'
resp = await fetch(URL)
boston_url = io.BytesIO((await resp.arrayBuffer()).to_py())
```

```python
boston_df=pd.read_csv(boston_url)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.boxplot(y=boston_df['MEDV'])
plt.title('Boxplot of Median Value of Owner-Occupied Homes')
plt.ylabel('MEDV ($1000s)')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='CHAS', data=boston_df)
plt.title('Bar Plot of Charles River Variable')
plt.xlabel('CHAS (1 = Near River, 0 = Not Near River)')
plt.ylabel('Count')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python
boston_df['AGE_GROUP'] = pd.cut(boston_df['AGE'],
                                bins=[0, 35, 70, 100],
                                labels=['<=35', '35~70', '>70'])

plt.figure(figsize=(6, 4))
sns.boxplot(x='AGE_GROUP', y='MEDV', data=boston_df)
plt.title('Boxplot of MEDV by Age Group')
plt.xlabel('Age Group')
plt.ylabel('MEDV ($1000s)')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x='INDUS', y='NOX', data=boston_df)
plt.title('NOX vs Proportion of Non-retail Business Acres')
plt.xlabel('INDUS (% non-retail business acres)')
plt.ylabel('NOX (Nitric Oxide Concentration)')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(boston_df['PTRATIO'], bins=15, kde=True)
plt.title('Histogram of Pupil-Teacher Ratio')
plt.xlabel('PTRATIO')
plt.ylabel('Frequency')
plt.show()
```
